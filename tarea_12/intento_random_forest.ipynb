```python
list_year_x = list()
list_mount_x = list()
list_day_x = list()
list_year_y = list()
list_mount_y = list()
list_dayu_y = list()

for index, row in df_sample.iterrows():
    date_x = str(row['Date_x'])
    date_y = str(row['Date_y'])
    year_x = date_x.split("-")[0]
    mount_x =  date_x.split("-")[1]
    day_x =  date_x.split("-")[2]
    year_y = date_y.split("-")[0]
    mount_y =  date_y.split("-")[1]
    day_y =  date_y.split("-")[2]
    list_year_x.append(year_x)
    list_mount_x.append(mount_x)
    list_day_x.append(day_x[0:2])
    list_year_y.append(year_y)
    list_mount_y.append(mount_y)
    list_dayu_y.append(day_y[0:2])
     

new_df=dem_chumingo.assign(Year_x=list_year_x)
new_df=new_df.assign(Month_x=list_mount_x)
new_df=new_df.assign(Day_x=list_day_x)
new_df=new_df.assign(Year_y=list_year_y)
new_df=new_df.assign(Month_y=list_mount_y)
new_df=new_df.assign(Day_y=list_dayu_y)
```
